2. **Определите, какие переменные являются непрерывными, а какие дискретными:**

   - **Непрерывные переменные:** могут принимать любое значение в диапазоне.
   - **Дискретные переменные:** принимают только определенные значения (обычно целые числа).

   **Ваши ответы:**

   - **Непрерывные переменные:**
     - age - возраст может быть любым, часто коругляется до целого
     -  resting blood pressure - артериальное давление в покое может варьироваться в широком диапазоне
     - serum cholestoral in mg/dl — уровень холестерина в крови также принимает любые значения
     - maximum heart rate achieved — максимальная частота сердечных сокращений может варьироваться в разных диапозонах
     - oldpeak — это показатель депрессии сегмента ST, который может принимать различные дробные значения ⏰
   
   - **Дискретные переменные:**
     - sex - принимает либо значение 1(муж), либо 0(жен)
     - chest pain type - принимает значение от 1 до 4-ёх, в зависимости от типа боли

     - slope of peak — наклон сегмента ST имеет три фиксированных значения, это дискретная переменная
     - number of major vessels — количество крупных сосудов с отклонениями, принимает четыре фиксированных значения
     - thal — результаты теста таллия, имеет три фиксированных значения
     - Disease — наличие сердечно-сосудистого заболевания (есть или нет) ⏰
